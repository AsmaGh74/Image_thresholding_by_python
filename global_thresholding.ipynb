```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# define a function to show the img

def img_show(img):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB) # we use open cv to read the image
    plt.figure(figsize=(10,7))
    plt.axis('off')
    plt.imshow(img, cmap = 'gray')
    plt.show()
```

```python
# read image

img_path = 'your/img/path/image_name.image_format'
img = cv.imread(img_path)
```

```python
# calculate img histogram

def img_histogram(img):
    gray =  cv.cvtColor(img, cv.COLOR_BGR2GRAY )
    img_show(gray)  # show gray scale image
    hist = cv.calcHist([gray],[0],None,[256],[0,256])
    plt.figure(figsize=(15,7))
    plt.grid()
    plt.plot(hist)
    plt.xticks(range(0,257,10))
    plt.show()
```

```python
# implement global thresholding

"""
discription:
if pixel gray level >= threshold: map pixel value to 255 (white)
and if pixel gray level < threshold: map pixel value to 0 (black)

inputs:
img: in BGR format (we use open cv to read the image)
threshold: threshold value
"""

def global_thresholding(img, threshold):
    gray =  cv.cvtColor(img, cv.COLOR_BGR2GRAY )
    img_show(gray)
    thresholded_img = np.zeros((gray.shape)).astype(np.uint8)
    indxs = np.where(gray >= threshold)
    for i,j in zip(indxs[0],indxs[1]):
        thresholded_img[i,j] = 255
    img_show(thresholded_img)
```

```python
# plot img histogram and choose threshold value based on it

img_histogram(img)
```

```python
# threshold image and show result

th = 50 # selected threshold value
global_thresholding(img, th)
```
